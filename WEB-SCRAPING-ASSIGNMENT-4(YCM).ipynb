# 4. Scrape the details of State-wise GDP of India from statisticstime.com.


```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)

page = requests.get(url)
url="http://www.imdb.com/list/ls095964455/"

driver.get(url)
```

```python
soup = BeautifulSoup(page.content)
```

```python
series_Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
```

```python
Name= soup.find_all("h3", class_="lister-item-header")
series_Name=[]
for i in Name:
    for j in i.find_all("a"):
        series_Name.append(j.text)
        
```

```python
year = soup.find_all("span",class_="lister-item-year text-muted unbold")
Year_span = []
for k in year:
    a=k.text.replace('(','')
    Year_span.append(a.replace(')','')) 
```

```python
Genres = soup.find_all("span",class_="genre")
Genre = []
for k in Genres:
    a=k.text
    Genre.append(a.replace('\n','')) 
```

```python
times = soup.find_all("span",class_="runtime")
Run_time = []
for k in times:
    a=k.text
    Run_time.append(a) 
```

```python
Rating = soup.find_all("span",class_="ipl-rating-star__rating")
Ratings = []
for k in Rating:
    a=k.text.replace('','')
    Ratings.append(a.replace('Rate','')) 
```

```python
Ratings
```

```python
Vote= driver.find_elements_by_xpath("/html/body/div[3]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div[1]/div[2]/p[4]/span[2]")
Votes = []
for k in Vote:
    a=k.text
    Votes.append(a) 
```

```python
Votes 
```

```python
print(len(series_Name),len(Year_span),len(Genre),len(Run_time),len(Ratings))
```

```python
df=pd.DataFrame({"series_Name":series_Name,"Year_span":Year_span,"Genre":Genre,"Run_time":Run_time})
df
```

# 7. Scrape the details of Data science recruiters from naukri.com.

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)

url="http://www.naukri.com/"

driver.get(url)

search_field_designation=driver.find_element_by_class_name("suggestor-input ")
search_field_designation.send_keys("Data Analyst")

search_button=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[6]")
search_button.click()
```

```python
Designations=[]
Companys=[]
Skills=[]
Locations=[]
```

```python
Designation_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")


for i in Designation_tags:
    Designation=i.text
    Designations.append(Designation)


Company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")


for i in Company_tags:
    Company=i.text
    Companys.append(Company)


Skill_tags=driver.find_elements_by_xpath("//ul[@class='tags has-description']")


for i in Skill_tags:
    Skill=i.text
    Skills.append(Skill)


location=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
for i in location:
    if i.text is None :
        Locations.append("--") 
    else:
        Locations.append(i.text)
        
```

```python
df=pd.DataFrame({"Companys":Companys[0:10],"Designations":Designations[0:10],"Skills":Skills[0:10],"Locations":Locations[0:10]})
df
```
